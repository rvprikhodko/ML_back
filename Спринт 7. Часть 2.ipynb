### Линейная регрессия с регуляризацией (Ridge, Lasso)

Использование регуляризации позволяет в процессе минимизации функции ошибки не допускать больших значений у коэффициентов (весов) модели. Использование регуляризаторов позволяет избежать переобучения моделей.

В рассматриваемых моделях значение коэффициента регуляризации ($\alpha$) по умолчанию равно 1.

```python
# обучим модель с L1-регуляризацией
ridge = Ridge().fit(X_regression_train, y_regression_train)
print(ridge.score(X_regression_test, y_regression_test))
print(mean_absolute_error(y_regression_test, ridge.predict(X_regression_test)))
print(mean_squared_error(y_regression_test, ridge.predict(X_regression_test)))
```

```python
lasso = Lasso().fit(X_regression_train, y_regression_train)
print(lasso.score(X_regression_test, y_regression_test))
print(mean_absolute_error(y_regression_test, lasso.predict(X_regression_test)))
print(mean_squared_error(y_regression_test, lasso.predict(X_regression_test)))
```

Можем увидеть, что при параметре $\alpha$ по умолчанию, метрики качества ухудшились. На практике, нам часто необходимо подбирать оптимальные параметры модели для достижения наивысшего качества. В каком-то смысле - это и есть основная работа датасатаниста!

На примере подбора параметра $\alpha$ давайте расссмотрим, как подбирать гиперпараметры моделей. У каждой модели в sklearn есть свои параметры (их можно посмотреть в документации). Все будет аналогично

```python
# импортируем класс GridSearchCV
# он позволит перебрать все комбинации параметров (значения для перебора мы зададим) и выбрать лучшую модель
from sklearn.model_selection import GridSearchCV
```

```python
# создаем словарь, ключами которого являются названия параметров (напоминаю, их можно в документации посмотреть)
# а значениями - список из возможных значений (которые будут перебираться)
parameters = {'alpha': [0.25, 0.5, 0.75, 1]}
```

Варнинг - не ошибка :) Видимо, что при $\alpha$ = 0.25 должно достигаться наилучшее качество модели Lasso. Проверим.

```python
print(lasso_optimal.score(X_regression_test, y_regression_test))
print(mean_absolute_error(y_regression_test, lasso_optimal.predict(X_regression_test)))
print(mean_squared_error(y_regression_test, lasso_optimal.predict(X_regression_test)))
```

Ну, конечно, не сильно лучше, но лучше. По крайней мере, мы теперь знаем, как подбирать параметры моделей.

### K ближайших соседей

Алгоритм KNN (k ближайших соседей) - это простейший метод классификации. Интересно то, что этот метод не требует какого-либо обучения. Он подразумевает хранение в себе тренировочных данных. Все расчеты проводятся при получении предсказания.

Итак, мы имеем обучающий набор данных. И хотим получить предсказание для нового объекта с помощью KNN. Что нужно сделать:
- определить значение параметра k (количество соседей)
- рассчитать расстояний между новым объектом (для которого хотим получить предсказание) и всеми остальными объектами из обучающей выборки (при этом расстояние рассчитывается также, как в школе считали расстояние между точками)
- когда у нас есть список из расстояний, мы сортируем их по возврастанию и выбираем первые k (то есть смотрим k ближайших объектов)
- чтобы получить предсказание, нам нужно просто выбрать класс, который чаще всего встречается у k ближайших соседей.

Постарайтесь понять идею алгоритма - он несложный и просто реализуется. Обращаю внимание еще раз на вычисление растояния. Формула как в школе. Пусть $x_{new}$ - объект, для которого получаем предсказание. А $x$ - это объект из обучающей выборки (напоминаю, что мы вычисляем расстояния для $x_{new}$ и всех объектов обучающей выборки).

Тогда формула для расстояний следующая: $d=\sqrt{Σ_{i=1}^{p}{(x_{new}^{i}-x^i)^2}}$, где p - количество признаков (столбцов в наборе данных, по которым мы предсказываем)

Одним из дополнительных заданий к этому спринту будет собственная реализация KNN. Пока будем использовать реализацию из KNN.

```python
# здесь импортируем сами модели
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, GradientBoostingClassifier

# это функции, которые помогут оценивать качество моделей
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_classification_test, knn.predict(X_classification_test)))
```

В этом отчете мы можем увидеть общий процент принятия верных решений - accuracy. При этом, есть и другие метрики - precision, recall и их производная метрика f1. Анализируя значения этих метрик видим, что объекты второго класса (1) предсказываются с большей точностью, чем объекты первого класса (1). Очевидная причина этому - преобладание объектов класса 1 в обучающем наборе данных. Посмотрим, как можно наглядно посмотреть, сколько объектов какого класса было предсказано.

## Глубокое обучение с Tensorflow

Глубокое обучение (deep learning) - это обучение глубоких нейронных сетей (глубокими нейронными сетями называют нейронные сети, состоящие из более чем одного скрытого слоя).
![mach-learn.png](https://neerc.ifmo.ru/wiki/images/thumb/6/63/Multi-layer-neural-net-scheme.png/500px-Multi-layer-neural-net-scheme.png)

Схема расчета значения одного нейрона (на основе значений нейронов предыдущего слоя).
![mach-learn.png](https://neerc.ifmo.ru/wiki/images/a/a5/%D0%98%D1%81%D0%BA%D1%83%D1%81%D1%81%D1%82%D0%B2%D0%B5%D0%BD%D0%BD%D1%8B%D0%B9_%D0%BD%D0%B5%D0%B9%D1%80%D0%BE%D0%BD_%D1%81%D1%85%D0%B5%D0%BC%D0%B0.png)

Предсказания нейронной сети - это значения нейронов выходного слоя (решая задачу классификации, мы размещаем на выходном слое 1 нейрон; решая задачу классификации - k нейронов, по количеству классов). Функции активации позволяют управлять множеством значений нейрона. Ниже приведены наиболее часто используемые функции активации.
![mach-learn.png](https://neurohive.io/wp-content/uploads/2018/07/funkcii-aktivacii-neironnoi-seti-570x257.png)

Обучаются нейронные сети методом обратного распространения ошибки. Он представляет собой оптимизацию (минимизацию) некоторой функции ошибки (loss-функции), которая вычисляется на основе полученных предсказаний и истинных значений. MSE, используемая как метрика в задаче регрессии, используется в качестве функции потерь при решении задачи регрессии. Для решения задачи классификации, как правило, используется категориальная кроссэнтропия (для случая двух классов - бинарная кроссэнтропия). Более подробно ознакомиться с этой функцией предлагаю самостоятельно.

Суть минимизации - вычисление антиградиента функции ошибки и выполнение шага градиентного спуска с целью обновления весов (градиент представляет собой вектор частных производных по всем весам). Основа Tensorflow (помимо структуры данных Тензор) - возможность эффективного автодифференцирования, с использованием ресурсов GPU.

Tensorflow автоматически по умолчанию использует GPU, если у вас GPU от NVidia и в системе установлены библиотеки CUDA и cuDNN.

Мы будем обучать сети без использования графического процессора.

### Решение задачи регрессии

```python
X_regression
```

На входном слое у нас должно быть 40 нейронов (потому что один дом характеризуется 40 признаками). На выходном слое должен быть один нейрон, потому что мы предсказываем одно чисило (стоимость дома). Функция активации выходного слоя не должна никак ограничивать значение нейрона. Скрытых слоев может быть сколько угодно, в них может быть сколько угодно нейронов - мы сами это определяем. Построем нейросеть с помощью API Keras фреймворка Tensorflow.

```python
model_regression = tf.keras.Sequential(
    [
        # Dense - полносвязный слой (каждый нейрон следующего слоя связан со всеми нейронами предыдущего)
        tf.keras.layers.Dense(64, activation="elu", input_shape=(40,)),
        # на втором скрытом слое будет 32 нейрона
        tf.keras.layers.Dense(32, activation="elu"),
        # Dropout позволяет внести фактор случайности - при обучении часть нейронов будет отключаться
        # каждый нейрон, в данном случае, будет отключаться с вероятностью 0.15
        tf.keras.layers.Dropout(0.15),
        tf.keras.layers.Dense(16, activation="elu"),
        tf.keras.layers.Dropout(0.2),
        # на выходе один нейрон
        tf.keras.layers.Dense(1, activation="linear"),
    ]
)
```

```python
# посмотрим, какая сеть у нас получилась
model_regression.summary()
```

Видим количество обучаемых параметров каждого слоя и общее количество обучаемых параметров. Перед использованием модель необходимо скомпилировать, при этом указывается оптимизатор, скорость обучения (можно представлять как величину шага в методе градиентного спуска), функция потерь и метрики, которые мы хотим (при желании) вычислять в будущем методом evaluate()

```python
model_regression.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="mse")
```

Мы готовы начинать обучения. В процессе обучения выключим логгирование (verbose=None). Основной параметр обучения - количеcтво эпох. Оно показывает, сколько раз модель пропустит через себя всю обучающую выборку. При этом существуют разные подходы к обучению - можно не просматривать всю выборку на одной эпохе, а генерировать каждый раз случайную подвыборку и обучаться на ней в рамках текущей эпохи. Это позволит сэкономить время обучения и предотвратить возможное переобучение.

```python
model_regression.fit(X_regression_train, y_regression_train, epochs=10, verbose=None)
```

Проверим нашу модель с помощью метрик, которые использовали ранее.

```python
print(mean_absolute_error(y_regression_test, model_regression.predict(X_regression_test)))
print(mean_squared_error(y_regression_test, model_regression.predict(X_regression_test)))
```

Видим, что результаты обучения есть, но они пока хуже моделей, которые мы рассматривали ранее. Попробуем изменить архитектуру сети и увеличить количество эпох.

```python
model_regression = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(64, activation="relu", input_shape=(40,)),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.05),
        tf.keras.layers.Dense(256, activation="relu"),
        tf.keras.layers.Dropout(0.025),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.05),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(32, activation="relu"),
        # на выходе один нейрон
        tf.keras.layers.Dense(1, activation="linear"),
    ]
)
model_regression.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss="mse")
model_regression.fit(X_regression_train, y_regression_train, epochs=50, verbose=None)
print(mean_absolute_error(y_regression_test, model_regression.predict(X_regression_test)))
print(mean_squared_error(y_regression_test, model_regression.predict(X_regression_test)))
```

Мы получили наглядную демонстрацию важного факта - в некоторых задачах применение нейронных сетей менее целесообразно, чем использование более простых моделей. Но иногда они дают лучшие результаты. Важную роль еще играет подбор архитектуры и параметров.

### Решение задачи классификации

```python
X_classification
```

Нейронная сеть для решения задачи классификации будет очень похожа на ту сеть для регрессии, однако у нее по другому будет организован выходной слой. У нас есть 2 стратегии наполнения выходного слоя нейронами:
- при решении задачи бинарной классификации мы можем расположить на выходном слое один нейрон с функцией активации sigmoid (значения от 0 и 1), после чего округлять полученные значения; значение нейрона покажет уверенность сети в предсказании; также мы можем расположить 2 нейрона на выходном слое и применить функцию softmax. Тогда сумма значений нейронов выходного слоя будет 1, а предсказание мы сможем получить определив нейрон с наибольшим значением;
- в случае многоклассовой классификации, как правило, на выходном слое располагаются k нейронов (по количеству классов), функция активации - softmax; нейрон с наибольшим значением определяет предсказанный класс.

У нас задача бинарной классификации, попробуем обе стратегии.

```python
model_classification_1 = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(64, activation="relu", input_shape=(37,)),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.05),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.Dense(16, activation="relu"),
        # сначала используем 1 нейрон и sigmoid
        tf.keras.layers.Dense(1, activation="sigmoid"),
    ]
)
# в качестве функции активации используется бинарная  кроссэнтропия
model_classification_1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="binary_crossentropy")
model_classification_1.fit(X_classification_train, y_classification_train, epochs=25, verbose=None)
```

```python
model_classification_1.predict(X_classification_test)[:5]
```

```python
y_pred = np.around(model_classification_1.predict(X_classification_test))
print(classification_report(y_classification_test, y_pred))
print(confusion_matrix(y_classification_test, y_pred))
```

Нейросеть отлично научилась выделять объекты более распространенного класса. Но в данных сильный дисбаланс и результаты отличаются от ожидаемых. Можем указать веса, которые будут использоваться при оптимизации функции ошибки. В качестве весов классов можно задать величины, обратные количеству элементов класса.

```python
w0 = 1 / y_classification_train[y_classification_train==0].shape[0]
w1 = 1 / y_classification_train[y_classification_train==1].shape[0]
```

```python
model_classification_1 = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(64, activation="relu", input_shape=(37,)),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.05),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.Dense(16, activation="relu"),
        # сначала используем 1 нейрон и sigmoid
        tf.keras.layers.Dense(1, activation="sigmoid"),
    ]
)
model_classification_1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="binary_crossentropy")
model_classification_1.fit(X_classification_train, y_classification_train, epochs=25, verbose=None,
                           class_weight={0: w0, 1: w1})
y_pred = np.around(model_classification_1.predict(X_classification_test))
print(classification_report(y_classification_test, y_pred))
print(confusion_matrix(y_classification_test, y_pred))
```

Видим улучшения. Можем поиграть с архитектурой и параметрами и добиться еще более качественных результатов. Но напоследок давайте попробуем разместить 2 нейрона на выходном слое и использовать softmax в качестве функции активации.

```python
model_classification_2 = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(64, activation="relu", input_shape=(37,)),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.05),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.Dense(16, activation="relu"),
        # сначала используем 2 нейрона и softmax
        tf.keras.layers.Dense(2, activation="softmax"),
    ]
)
# в качестве функции активации используется категориальная кроссэнтропия
# используем разряженный (sparse) вариант, поскольку значения целевого признака не закодированы One-Hot кодированием
model_classification_2.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="sparse_categorical_crossentropy")
model_classification_2.fit(X_classification_train, y_classification_train, epochs=25, verbose=None,
                           class_weight={0: w0, 1: w1})
```

```python
model_classification_2.predict(X_classification_test)
```

```python
model_classification_2.predict(X_classification_test)[:5]
```

```python
# получим индексы максимального значения для каждого элемента (вложенный массив) с помощью numpy
y_pred = [np.argmax(pred) for pred in model_classification_2.predict(X_classification_test)]
```

```python
print(classification_report(y_classification_test, y_pred))
print(confusion_matrix(y_classification_test, y_pred))
```
